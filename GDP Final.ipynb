```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
```

```python
gdp1=pd.read_csv(r'gdp_new1.csv')
gdp1
```

```python
gdp_data1=gdp1.drop(['Source:','Native Frequency:'],axis=1)
gdp_data1
```

```python
gdp_data2 =pd.melt(gdp_data1, id_vars=['Description:',  'Geography:'], var_name='Quarter', value_name='GDP')
gdp_data2
```

```python
gdp_data2=gdp_data2.drop([ 'Description:'], axis = 1)
gdp_data2
```

```python
df = gdp_data2.sort_values(by=[ 'Geography:','Quarter'])
df
```

```python
# Chart to update the plot based on selected geography
def plot(selected_geo):
    plt.figure(figsize=(12, 8))
    for geo in df['Geography:'].unique():
        data = df[df['Geography:'] == geo]
        if geo == selected_geo:
            plt.plot(data['Quarter'], data['GDP'], label=geo, linewidth=2, color='blue')
        else:
            plt.plot(data['Quarter'], data['GDP'], label=geo, linestyle='--', alpha=0.5)

    plt.xlabel('Quarter')
    plt.ylabel('GDP')
    plt.title(f'GDP by Geography over Time ({selected_geo} highlighted)')
    plt.legend()
    plt.show()

# Create an interactive widget
interact(plot, selected_geo=df['Geography:'].unique())
```

```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import pandas as pd
import re

# Step 1: Drop rows with invalid or missing values
df = df.dropna(subset=['Quarter', 'GDP', 'Geography:'])

# Step 2: Ensure correct data types
df['Geography:'] = df['Geography:'].astype(str)
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')
df = df.dropna(subset=['GDP'])

# Step 3: Clean and ensure 'Quarter' is in correct format (e.g., '1995Q1')
df['Quarter'] = df['Quarter'].astype(str)
quarter_pattern = re.compile(r'^\d{4}Q[1-4]$')
df = df[df['Quarter'].apply(lambda x: bool(quarter_pattern.match(x)))]

# Step 4: Convert Quarter to datetime for sorting
def quarter_to_date(q):
    year = int(q[:4])
    quarter = int(q[-1])
    return pd.Timestamp(f"{year}-{(quarter - 1) * 3 + 1:02d}-01")

df['QuarterDate'] = df['Quarter'].apply(quarter_to_date)
df = df.sort_values('QuarterDate')

# Chart to update the plot based on selected geography
def plot(selected_geo):
    plt.figure(figsize=(12, 8))
    for geo in df['Geography:'].unique():
        data = df[df['Geography:'] == geo]
        if geo == selected_geo:
            plt.plot(data['QuarterDate'], data['GDP'], label=geo, linewidth=2, color='blue')
        else:
            plt.plot(data['QuarterDate'], data['GDP'], label=geo, linestyle='--', alpha=0.5)

    plt.xlabel('Quarter')
    plt.ylabel('GDP')
    plt.title(f'GDP by Geography over Time ({selected_geo} highlighted)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Create an interactive widget
interact(plot, selected_geo=df['Geography:'].unique())
```

```python
df = df.dropna(subset=['Quarter', 'GDP', 'Geography:'])  # Remove rows with missing data

# Filter only valid Quarter values (e.g., 2020Q1)
quarter_pattern = re.compile(r'^\d{4}Q[1-4]$')
df = df[df['Quarter'].astype(str).apply(lambda x: bool(quarter_pattern.match(x)))]

# Convert types
df['Quarter'] = df['Quarter'].astype(str)
df['Geography:'] = df['Geography:'].astype(str)
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')
df = df.dropna(subset=['GDP'])

# Function to plot the chart
def plot(selected_geo):
    plt.figure(figsize=(12, 10))
    
    data = df[df['Geography:'] == selected_geo]
    plt.plot(data['QuarterDate'], data['GDP'], label=selected_geo, linewidth=2, color='salmon')

    plt.xlabel('Quarter')
    plt.ylabel('GDP')
    plt.title(f'GDP by Geography over Time ({selected_geo})')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()
    plt.show()

# Create an interactive widget
geo_dropdown = widgets.Dropdown(
    options=df['Geography:'].unique(),
    value=df['Geography:'].unique()[0],
    description='Geography:'
)

interact(plot, selected_geo=geo_dropdown)
```
